#Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
  
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
  
# fit the regressor with x and y data
regressor.fit(X_train,Y_train) 
prediction = regressor.predict(X_test)
```

```python
prediction
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score, r2_score,accuracy_score
from math import sqrt

mse = mean_squared_error(Y_test, prediction)
errors = abs(prediction - Y_test)
mape = 100 * (errors / Y_test)
accuracy = 100 - np.mean(mape)
regressor.score(X_test,Y_test)
print("Accuracy Score: ",round(accuracy,2))
print("Mean Squared Error: ",round(mse,2))
```

#Arima MOdel

```python
from statsmodels.tsa.arima.model import ARIMA as ARIMA
```

```python
# split into train and test sets
size = int(len(Y) * 0.77)
train, test = Y[0:size], Y[size:len(Y)]
history = [x for x in train]
predictions = list()
actual = list()
# walk-forward validation
for t,u in zip(range(len(test)),test):
  model = ARIMA(history, order=(5,1,0))
  model_fit = model.fit()
  output = model_fit.forecast()
  yhat = output[0]
  predictions.append(yhat)
  actual.append(u)
  obs = u
  history.append(obs)
  print('predicted=%f, expected=%f' % (yhat, obs))
```

```python
# Summary of model
print(model_fit.summary())
```

```python
from matplotlib import pyplot
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error

from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
#Mean Absolute Error (MAE)
print("\n\nMean Absolute Error (MAE) -: ",mean_absolute_error(actual,predictions))
#Mean Absolute percentage Error (MAPE)
print("Mean Absolute percentage Error (MAPE) -: ",mean_absolute_percentage_error(actual,predictions))
# Mean square error (MSE)
print("Mean square error (MSE) -: ",mean_squared_error(actual,predictions))
# Root mean squared error (RMSE)
rmse = sqrt(mean_squared_error(actual,predictions))
print('Root mean squared error (RMSE): %.3f' % rmse)
```

```python
# plot predicted values against actual outcomes

pyplot.plot(actual)
pyplot.plot(predictions, color='red')
plt.title('Line Plot of Actual and Predicted values of ARIMA')
plt. xlabel("Count")
plt.ylabel("Values")
pyplot.show()
```

#LSTM Model

```python
from sklearn.metrics import mean_absolute_error
from keras.models import Sequential
from sklearn import metrics
import tensorflow as tf
from keras.layers import Dense, LSTM, Dropout,Flatten

model = Sequential()
model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(optimizer='adam',loss='mean_squared_error',metrics=[tf.keras.metrics.CategoricalAccuracy()])
history=model.fit(X_train,Y_train,epochs=5,batch_size=32)
```
